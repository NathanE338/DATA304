```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("system_of_record_timestamps.csv")
df
```

```python
df['Primary_Timestamp'] = pd.to_datetime(df['Primary_Timestamp'], infer_datetime_format=True,utc=True)

# Format all dates to a single format (e.g., YYYY-MM-DD)
df['Primary_Timestamp'] = df['Primary_Timestamp'].dt.strftime('%Y-%m-%d')
df
# Making all the dates in the Primary_Timestamp column into one format.
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Primary_Timestamp'], df['Event_Count'].cumsum(), marker='o', linestyle='-', color='b')
plt.title('Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Event Count')
plt.grid(True)
plt.show()
#Plotting the dates and event count to show that the timestamp ins continuous, thus the problem with the server
```

Using the graph above, we see that the times are logged continously. There are other clues that lead us to this conclusion, such as the NaN values in the csv file.
